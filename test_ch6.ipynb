```python
import pandas as pd

df = pd.DataFrame(data={'topic':['international','sports','sports','politics', 'international'],
                        'content':['test2','testing','this is test','x','df']})
df
```

```python
from sklearn import preprocessing

encoder= preprocessing.LabelEncoder()
encoder.fit(df['topic'])
df['topic'] = encoder.fit_transform(df['topic'])
df
```

```python
print(encoder)
encoder.classes_

```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# df already exists with column 'topic'

encoder = OneHotEncoder(sparse_output=False)  # or sparse=False in older versions

# OneHotEncoder expects 2D input, so use double brackets
topic_oh = encoder.fit_transform(df[['topic']])

# Make a DataFrame from the result
topic_oh_df = pd.DataFrame(
    topic_oh,
    columns=encoder.get_feature_names_out(['topic']),
    index=df.index
)

# Concatenate with original df, and optionally drop original 'topic'
df_oh = pd.concat([df.drop(columns=['topic']), topic_oh_df], axis=1)
df_oh
```
